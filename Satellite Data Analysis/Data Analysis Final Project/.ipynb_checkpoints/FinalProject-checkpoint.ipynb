```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
from scipy.stats import norm, kde
import math

df = pd.read_csv("UCS-Satellite-Database-1-1-2022.xlsx - Sheet1.csv")
```

```python
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
df = df.loc[:, ~df.columns.str.contains('^Source')]
df = df.loc[:, ~df.columns.str.contains('Number')]
df = df.loc[:, ~df.columns.str.contains('Comments')]
df = df.loc[:, ~df.columns.str.contains('Detailed Purpose')]
df = df.loc[:, ~df.columns.str.contains('^Longitude of GEO')]
df.dropna(subset = ["Users"], inplace=True)
dfNotClean = df.copy()
df= df.dropna()
```

```python
purpose = df['Purpose']
perigee = df['Perigee (km)']
apogee = df['Apogee (km)']
eccentricity = df['Eccentricity']
period = df['Period (minutes)']
launchMass = df['Launch Mass (kg.)']
df = df.replace(',','', regex=True)
df['Perigee (km)'].astype(int)
df.sort_values('Perigee (km)')
df.index = range(len(df))
```

As perigee (point where distance is closest to earth) and apogee (point where distance is farthest to earth) increase, period also increases due to some science stuff
launch mass also seems to increase as the distance increases, purpose is probably different too

```python
data = [df['Perigee (km)'].astype(int),df['Apogee (km)'].astype(int),df['Eccentricity'],df['Period (minutes)'],df['Launch Mass (kg.)'].astype(int)]
headers = ['Perigee (km)', 'Apogee (km)','Eccentricity','Period (minutes)','Launch Mass (kg.)']
spaceData = pd.concat(data, axis=1, keys=headers)

spaceData = spaceData[spaceData['Perigee (km)']<800]
spaceDataNorm = spaceData.copy()



for column in spaceDataNorm.columns:
    spaceDataNorm[column] = spaceDataNorm[column]  / spaceDataNorm[column].abs().max()
spaceData.describe()
```

```python
spaceDataNorm.describe()
```

```python
plt.scatter(spaceDataNorm['Perigee (km)'],spaceDataNorm['Period (minutes)'], alpha= 0.5)
plt.ylabel('Normalized Apogee Level')
plt.xlabel('Normalized Period')
plt.title('Spread of Normalized Period vs. Normalized Apogee')
plt.show()
```

```python
x = spaceDataNorm['Perigee (km)']
y = spaceDataNorm['Period (minutes)']

nbins=300
k = kde.gaussian_kde([x,y])
xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))

plt.pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto')
plt.ylabel('Normalized Apogee Level')
plt.xlabel('Normalized Period')
plt.title('Density of Normalized Period vs. Normalized Apogee')
plt.show()
```

```python
bars = ("Civil", "Commercial", "Government", "Military")
height = vals['Launch Mass (kg.)']
y_pos = np.arange(len(bars))
plt.bar(y_pos, height)
plt.xticks(y_pos, bars)
plt.xlabel('Uses')
plt.ylabel('Average Normalized Launch Mass')
plt.title('Launch Mass For Each Use')
plt.show()
```

```python
pCiv = vals['Perigee (km)']['Civil']
pCom = vals['Perigee (km)']['Commercial']
pGov = vals['Perigee (km)']['Government']
pMil = vals['Perigee (km)']['Military']
pbars = {pCiv, pCom, pGov, pMil}

aCiv = vals['Apogee (km)']['Civil']
aCom = vals['Apogee (km)']['Commercial']
aGov = vals['Apogee (km)']['Government']
aMil = vals['Apogee (km)']['Military']
abars = {aCiv, aCom, aGov, aMil}

eCiv = vals['Eccentricity']['Civil']
eCom = vals['Eccentricity']['Commercial']
eGov = vals['Eccentricity']['Government']
eMil = vals['Eccentricity']['Military']
ebars = {eCiv, eCom, eGov, eMil}

perCiv = vals['Period (minutes)']['Civil']
perCom = vals['Period (minutes)']['Commercial']
perGov = vals['Period (minutes)']['Government']
perMil = vals['Period (minutes)']['Military']
perbars = {perCiv, perCom, perGov, perMil}


barWidth = 0.15
r1 = np.arange(len(pbars))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]

plt.bar(r1, pbars, color='#ff3300', width=barWidth, edgecolor='white', label='Perigee')
plt.bar(r2, abars, color='#0026ff', width=barWidth, edgecolor='white', label='Apogee')
plt.bar(r3, ebars, color='#8c00ff', width=barWidth, edgecolor='white', label='Eccentricity')
plt.bar(r4, perbars, color='#ffffff', width=barWidth, edgecolor='black', label='Period')

plt.xlabel('Uses')
plt.ylabel('Normalized Average Value')
plt.title('Normalized Orbit Data For Each Use')
plt.xticks([r + barWidth for r in range(len(abars))], ['Civil', 'Military', 'Commercial', 'Government'])
plt.legend(bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
keepCount=0
for index, row in dfNotClean.iterrows():
    if row['Country/Org of UN Registry']==row['Country of Operator/Owner']:
        keepCount = keepCount+1
        
nKeepCount=0
for index, row in dfNotClean.iterrows():
    if row['Country/Org of UN Registry']!=row['Country of Operator/Owner']:
        nKeepCount = nKeepCount+1

values = [keepCount, nKeepCount]
colors = ['#ebafa9', '#ada9eb']
fig1, ax1 = plt.subplots(figsize=(10,5))
plt.pie(values, autopct='%1.1f%%', labeldistance=1.15, wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' }, colors=colors);
plt.title('Percentage of Satellites Kept by Country of Origin')
plt.legend(labels=['Kept', 'Not Kept'], loc=3)
ax1.axis('equal')
plt.show()
```
